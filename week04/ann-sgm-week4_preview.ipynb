1. Data Processing

```python
import tensorflow
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
new_data = []
for data in train_data['Sex']:
    if data == 'male':
        new_data.append(0)
    elif data == 'female':
        new_data.append(1)
    else:
        new_data.append(2)
train_data['Sex'] = new_data
print(train_data)
```

1-2. nan 찾기 (dropna() 사용)

```python
proc_data = train_data.dropna()
print(proc_data.isnull().sum())
print(proc_data)
```

1-2. nan찾기(다른방법)

```python
new_data = []
for label in train_data:
    print(label)
    for index, data in zip(train_data.index, train_data[label]):
        if str(data) == str(float('nan')):
            new_data = train_data.drop(index, axis=0)
            print(index)
print(train_data)
print(new_data)
```
