Step1. 

```python
from tensorflow.keras import layers, models
```

Step2. Hyper-params 설정 (Nin, Nh, Nout 같은 것들)

Step3. 실제 구현
-Dense(layer 개수, activation func, input_shape)
-input_shape가 생략될 경우, 앞 layer의 출력 개수가 들어간다.
-compile

```python
def ANN_seq_func(Nin, Nh, Nout):
  model = models.Sequential()
  model.add(layers.Dense(Nh, activation='relu', input_shape=(Nin,)))
  model.add(layers.Dense(Nout, activation='softmax'))
  model.compile(loss='categorical_crossentropy',
                optimizer='rmsprop', metrics=['accuracy'])
  return model
```

Step4. load and preprocess data

```python
from tensorflow.keras.datasets import mnist
#data load
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

#data reshape and normalization
L, W, H = train_images.shape
train_images = train_images.reshape((60000, W*H))
train_images = train_images.astype('float32')/255
test_images = test_images.reshape((10000, W*H))
test_images = test_images.astype('float')/255

#binary target
from tensorflow.keras.utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

Step 5&6. Training and Evaluation

```python
#fitting
network = ANN_seq_func(W*H,512,10)
network.fit(train_images, train_labels, epochs=10, batch_size=128)

#evaluate on test set
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc:', test_acc)
```
