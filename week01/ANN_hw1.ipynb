numpy의 메소드 sum을 이용한 풀이

```python
import numpy as np

data = np.loadtxt("/content/drive/MyDrive/Colab Notebooks/week1/practice1.csv", delimiter=",")
data_sum = np.sum(data)
print(data_sum)
```

for문을 이용한 풀이

```python
import numpy as np

data = np.loadtxt("/content/drive/MyDrive/Colab Notebooks/week1/practice1.csv", delimiter=",")
data_sum = 0
for d in data:
  data_sum = data_sum + d
print(data_sum)
```
